### Data Preparation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt     
from pandas.plotting import lag_plot

%matplotlib inline 
```

```python
data = pd.read_csv("/Users/angelazhao/Desktop/ST306/Homework/HW1/hw1.csv")
```

```python
df = pd.DataFrame(data)
```

```python
df.set_index(["Date"], inplace = True)
```

### Problem 1: Calculate linear and compounded returns

```python
lin_rtn = df.pct_change().iloc[1:, :]
lin_rtn
```

```python
com_rtn = np.log(df / df.shift(1)).iloc[1:, :]
com_rtn
```

### Problem 2: Visual Test

```python
def f(x):
    fig, axes = plt.subplots(1, 3, figsize = (18, 5))
    fig.suptitle(pd.DataFrame(x).columns)
    
    plt.subplot(131)
    x[:2258].hist()
    
    plt.subplot(132)
    x[2258:].hist()
    
    plt.subplot(133)
    lag_plot(x)
```

```python
lin_rtn.apply(f)
```

In conclusion, the linear returns for SPGSCITR Index and LF98TRUU Index are invariants because the histograms are symmetrical and the scatter plots are circular clouds. 

```python
com_rtn.apply(f)
```

In conclusion, the compounded returns for LF98TRUU Index is an invariant because the histograms are symmetrical and the scatter plot is a circular cloud.
